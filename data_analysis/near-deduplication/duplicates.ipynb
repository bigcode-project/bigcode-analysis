```python
import os

import pandas as pd
from datasets import load_dataset, load_from_disk
from IPython.display import display, Markdown
from rich.box import HEAVY
from rich.table import Table
from rich import print
```

```python
# load the indexed dataset
ds = load_from_disk("results/bigcode/the-stack-dedup-pjj/python/85/data/python/train/content/indexed")
neighbors = load_from_disk("results/bigcode/the-stack-dedup-pjj/python/85/data/python/train/content/neighbors")
```

```python
assert len(ds) == len(neighbors), f"Mismatched dataset sizes: {len(ds)} VS {len(neighbors)}"
```

```python
%%capture --no-display

duplicates = neighbors.filter(lambda x: x["__neighbors__"], num_proc=20)
```

```python
samples = duplicates.select(range(10))
display(Markdown(pd.DataFrame(samples)[["__id__", "__neighbors__"]].to_markdown()))
```

```python
import re

def jaccard_similarity(x, y):
    a = set(re.split("[^A-Za-z_0-9]", x))
    b = set(re.split("[^A-Za-z_0-9]", y))

    return len(a.intersection(b)) / len(a.union(b))

jaccard_similarity("hello world", "hello")
```

```python
import random

selection, selecttion_duplicates = samples[4]["__id__"], samples[4]["__neighbors__"]
selection_content = ds[selection]["content"]

for dup_id in random.sample(selecttion_duplicates, 3):
    
    table = Table(show_header=True, header_style="bold magenta", box=HEAVY)
    table.add_column("Selection", style="green")
    table.add_column("Duplicate", style="red")
    table.add_column("Actual Jaccard Similarity", style="cyan")
    table.add_row(
        selection_content, 
        ds[dup_id]["content"],
        f'{jaccard_similarity(selection_content, ds[dup_id]["content"]):.2f}'
    )

    print(table)
```
